```python
import joblib
import pandas as pd
```

```python


# Load model
model = joblib.load("WeatherAI.joblib")

print("==== Informasi Model ====")

# Jika model mendukung feature_names_in_
if hasattr(model, "feature_names_in_"):
    print("\n📌 Feature yang digunakan saat training:")
    for f in model.feature_names_in_:
        print(" -", f)
else:
    print("\n⚠️ Model tidak menyimpan nama feature (kemungkinan model lama).")

# Coba prediksi dummy untuk melihat jumlah fitur
try:
    n_features = model.n_features_in_
    print(f"\n📌 Jumlah fitur yang dibutuhkan model: {n_features}")
except:
    print("\n⚠️ Tidak dapat membaca jumlah fitur.")

print("\nSelesai.")

print(model.feature_names_in_)
```
