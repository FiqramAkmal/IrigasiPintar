```python
import requests
import pandas as pd
from datetime import datetime
```

```python
url = "https://api.open-meteo.com/v1/forecast"
params = {
    "latitude": -6.926729024500985,
    "longitude":  107.77489652305427,
    "minutely_15": "precipitation,temperature_2m,relative_humidity_2m",
    "timezone": "Asia/Bangkok",
    "past_days": 1,
    "forecast_days": 1,
    "forecast_minutely_15": 96,
	"past_minutely_15": 96
}
```

```python
res = requests.get(url, params=params).json()

minutely = res["minutely_15"]

df = pd.DataFrame({
    "date": pd.to_datetime(minutely["time"]),
    "precipitation": minutely["precipitation"],
    "temperature_2m": minutely["temperature_2m"],
    "relative_humidity_2m": minutely["relative_humidity_2m"]
})

# ambil waktu saat ini
#now = pd.to_datetime(datetime.now())

# filter data mulai jam sekarang (yang >= current time)
# df_filtered = df[df["date"] >= now].reset_index(drop=True)

# print(df_filtered)

df.to_csv("weather_15minutely.csv", index=False)
print("\nCSV disimpan → weather_hourly.csv")
```
